```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from Excel file
df = pd.read_excel("/Users/farrelmanazilin/Downloads/data/gemas.xlsx")
# Plot the time series
plt.figure(figsize=(10, 6))

# Plot Consumption Goods
plt.plot(df.index, df['Consumption Goods'], label='Consumption Goods')

# Plot Raw Material Support
plt.plot(df.index, df['Raw Material Support'], label='Raw Material Support')

# Plot Capital Goods
plt.plot(df.index, df['Capital Goods'], label='Capital Goods')

# Customize the plot
plt.xlabel('Tahun')
plt.ylabel('Value')
plt.title('Import Plot')
plt.legend()

# Display the plot
plt.show()
```

```python
# Plot the time series
plt.figure(figsize=(10, 6))

# Plot MIGAS
plt.plot(df.index, df['MIGAS'], label='MIGAS')

# Plot NON MIGAS
plt.plot(df.index, df['NON MIGAS'], label='NON MIGAS')

# Plot Agriculture
plt.plot(df.index, df['Agriculture'], label='Agriculture')

# Plot Industry
plt.plot(df.index, df['Industry'], label='Industry')

# Plot Mining
plt.plot(df.index, df['Mining'], label='Mining')

# Plot Others
plt.plot(df.index, df['Others'], label='Others')

# Customize the plot
plt.xlabel('Tahun')
plt.ylabel('Value')
plt.title('Eksport Plot')
plt.legend()

# Display the plot
plt.show()
```
