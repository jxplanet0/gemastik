```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Read the data from Excel file
df = pd.read_excel("/Users/farrelmanazilin/Downloads/data/gemas.xlsx")

# Extract the independent variable (X) and dependent variable (y)
X = df.index.values.reshape(-1, 1)  # Use the index as the independent variable
y = df['Consumption Goods']  # Use 'Consumption Goods' as the dependent variable

# Create and fit the linear regression model
regression_model = LinearRegression()
regression_model.fit(X, y)

# Get the coefficient and intercept of the linear regression model
coefficient = regression_model.coef_[0]
intercept = regression_model.intercept_

# Make predictions using the linear regression model
y_pred = regression_model.predict(X)

# Calculate the R-squared score
r_squared = r2_score(y, y_pred)

# Calculate the number of observations and number of features
n_obs = len(X)
n_features = regression_model.coef_.shape[0]

# Calculate the Adjusted R-squared score
adjusted_r_squared = 1 - (1 - r_squared) * (n_obs - 1) / (n_obs - n_features - 1)

# Calculate the residuals
residuals = y - y_pred

# Calculate the standard error (SE)
se = (mean_squared_error(y, y_pred)) ** 0.5

# Calculate the root mean square error (RMSE)
rmse = mean_squared_error(y, y_pred, squared=False)

# Calculate the mean absolute error (MAE)
mae = mean_absolute_error(y, y_pred)

# Calculate the root relative squared error (RRSE)
rrse = (rmse / (y.max() - y.min())) * 100

# Calculate the relative absolute error (RAE)
rae = (mae / (y.max() - y.min())) * 100

# Print the evaluation metrics
print("R-squared (R²):", r_squared)
print("Adjusted R-squared (R²):", adjusted_r_squared)
print("Residuals:", residuals)
print("Standard Error (SE):", se)
print("Root Mean Square Error (RMSE):", rmse)
print("Mean Absolute Error (MAE):", mae)
print("Root Relative Squared Error (RRSE):", rrse)
print("Relative Absolute Error (RAE):", rae)
print("Coefficient:", coefficient)
print("Intercept:", intercept)

# Plot the regression line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred, color='red', linewidth=2, label='Linear Regression')
plt.xlabel('Tahun')
plt.ylabel('Value')
plt.title('Linear Regression Import Consumption Goods')
plt.legend()
plt.show()
```
